## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_df = original_df.loc[:,['Gender', 'SN', 'Age']]

unique_df = unique_df.drop_duplicates()

unique_df.head()                            
```

```python
gender_counts = original_df.groupby('Gender')

unique_members = unique_df['Gender'].value_counts()

percent_players = unique_members / total_players

gender_df = pd.DataFrame({"Total Counts": unique_members, "Percentage Players": (percent_players)})

gender_df["Percentage Players"] = gender_df["Percentage Players"].map("{:,.2%}".format)

gender_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchases = original_df.groupby(["Gender"]).count()["Price"].rename("Total Purchases")
purchase_avg = original_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
total_value = original_df.groupby(["Gender"]).sum()["Price"].rename("Purchase Count")

per_purchase = purchases / gender_df["Total Counts"]
```

```python
new_df = pd.DataFrame({"Purchase Count": purchases, 
                         "Average Purchase Price": purchase_avg,
                         "Total Purchase Value": total_value,
                         "Avg Total Purchase per Person": per_purchase,
                        }) 

new_df["Average Purchase Price"] = new_df["Average Purchase Price"].map("${:,.2f}".format)
new_df["Total Purchase Value"] = new_df["Total Purchase Value"].map("${:,.2f}".format)
new_df["Avg Total Purchase per Person"] = new_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

new_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,9999]

group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
unique_df["Range"] = pd.cut(unique_df["Age"],bins,labels=group_labels)

unique_df
```

```python
range_count = unique_df['Range'].value_counts()

age_group_percent = range_count / total_players

age_df = pd.DataFrame({"Total Count": range_count, "Percentage Players": age_group_percent})

age_df["Percentage Players"] = age_df["Percentage Players"].map("{:,.2%}".format)

age_df = age_df.sort_index()

age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
original_df["Range"] = pd.cut(unique_df["Age"],bins,labels=group_labels)

original_df
```

```python
price_total = original_df.groupby(["Range"]).sum()["Price"].rename("Total Purchase Value")
price_avg = original_df.groupby(["Range"]).mean()["Price"].rename("Average Purchase Price")
age_counts = original_df.groupby(["Range"]).count()["Price"].rename("Purchase Count")

per_purchase = price_total / age_df["Total Count"]

final_df = pd.DataFrame({"Purchase Count": age_counts, 
                         "Average Purchase Price": price_avg,
                         "Total Purchase Value": price_total,
                         "Avg Total Purchase per Person": per_purchase,
                        }) 

final_df["Avg Total Purchase per Person"] = final_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
final_df["Average Purchase Price"] = final_df["Average Purchase Price"].map("${:,.2f}".format)
final_df["Total Purchase Value"] = final_df["Total Purchase Value"].map("${:,.2f}".format)

final_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_counts = original_df.groupby(["SN"]).count()["Price"].rename("Purchase Count")
average_purchase = original_df.groupby(["SN"]).mean()["Price"].rename("Avg Purchase")
total_value = original_df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

top_spender_df = pd.DataFrame({"Purchase Count": purchase_counts, 
                         "Avg Purchase": average_purchase,
                         "Total Purchase Value": total_value
                        })

top_spender_df = top_spender_df.sort_values(["Total Purchase Value"], ascending=False)

top_spender_df["Avg Purchase"] = top_spender_df["Avg Purchase"].map("${:,.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:,.2f}".format)


top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = original_df.loc[:,['Item ID', 'Item Name', 'Price']]

popular_df
```

```python
purchase_counts = popular_df.groupby(["Item ID"]).count()["Price"].rename("Purchase Count")
total_purchase_value = popular_df.groupby(["Item ID"]).sum()["Price"].rename("Total Purchase Value")
price = popular_df.groupby(["Item ID"]).mean()["Price"].rename("Item Price")
item_name = popular_df["Item Name"]


item_df = pd.DataFrame({"Purchase Count": purchase_counts, 
                         "Item Price": price,
                         "Total Purchase Value": total_purchase_value,
                        "Item Name": item_name})

item_df = item_df.sort_values(["Purchase Count"], ascending=False)

item_df["Item Price"] = item_df["Item Price"].map("${:,.2f}".format)
item_df["Total Purchase Value"] = item_df["Total Purchase Value"].map("${:,.2f}".format)

item_df.head()
```
